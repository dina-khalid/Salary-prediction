```python
def specializeCategories(categories,cutoff):
    '''
    A function that takes categories and cutoff number then 
    return the categories which have elements more than the cutoff num
    and make the other categories in category "other"
    '''
    catigorical_map={}
    for i in range (len(categories)):
        if categories.values[i]>= cutoff:
            catigorical_map[categories.index[i]]=categories.index[i]
        else:
            catigorical_map[categories.index[i]]="other"
    return catigorical_map
```

```python
countryMap=specializeCategories(df.Country.value_counts(),450)
```

```python
df["Country"]=df["Country"].map(countryMap)
df.Country.value_counts()
```

```python
def clean_edlevel(x):
    '''
    Takes education level and return cleaned categories
    '''

    if x=='Masterâ€™s degree (M.A., M.S., M.Eng., MBA, etc.)':
        return 'Masterâ€™s degree'
    if x=='Bachelorâ€™s degree (B.A., B.S., B.Eng., etc.)' :
        return "Bachelorâ€™s degree"
    if x=='Other doctoral degree (Ph.D., Ed.D., etc.)'  or x=='Professional degree (JD, MD, etc.)' :
        return 'post grad'
    return "less than a Bachelor"
df['EdLevel']=df['EdLevel'].apply(clean_edlevel)
```

```python
x=df.drop('Salary',axis=1)
y=df['Salary']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1, random_state = 10)
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.ensemble import RandomForestRegressor
random_forest_reg = RandomForestRegressor(random_state=0)
random_forest_reg.fit(x_train, y_train.values)
```

```python
y_pred2 = random_forest_reg.predict(x_test)
error=np.sqrt(mean_squared_error(y_test,y_pred2))
print("${:,.02f}".format(error))
```

```python
from sklearn.model_selection import GridSearchCV

max_depth = [None, 2,4,6,8,10,12,14]
parameters = {"max_depth": max_depth}

regressor = RandomForestRegressor(random_state=0)
gs = GridSearchCV(regressor, parameters, scoring='neg_mean_squared_error')
gs.fit(x_train, y_train.values)
```

```python
regressor = gs.best_estimator_

regressor.fit(x_train, y_train.values)
y_pred = regressor.predict(x_test)
error = np.sqrt(mean_squared_error(y_test, y_pred))
print("${:,.02f}".format(error))
```
